```python
# Импорт необходимых инструментов
from mnist import MNIST
import numpy as np
import matplotlib.pyplot as plt
import sys
mnist_data = MNIST('./MNIST_DATA')
```

```python
# np.set_printoptions(threshold=sys.maxsize)
# np.set_printoptions(threshold=10)
```

```python
# Функция активации
def activate_function(current_value):
    if(current_value > 0):
        return 1
    else:
        return 0
```

```python
# Скорость обучения
learning_speed = 0.5
# Cдвиг
learning_offset = np.random.rand(1, 10)
# Эпохи обучения
max_epochs = 20
print(learning_offset)
```

```python
images_train, labels_train = mnist_data.load_training()

learning_data = mnist_data.process_images_to_numpy(images_train)

correct_answers = mnist_data.process_labels(labels_train)
```

```python
# Правильные ответы
y = correct_answers
# Нейроны
x = np.around(learning_data / 255)
x = np.hstack((x, np.ones((len(x), 1))))
# Обучающая выборка 60000 x 785
print(x)
```

```python
# Веса
w = np.random.rand(784, 10)
w = np.vstack((w, learning_offset))
# Веса на входы 785 x 10
print(w)
```

```python
temp = np.eye(10)
make_answer = np.vectorize(activate_function)

for epoch in range(max_epochs):

    for i in range(len(x)):
        # 1x10
        y_1 = x[i]@w
        y_1 = make_answer(y_1)
        ei = temp[y[i]] - y_1
        Di = learning_speed * np.dot(x[i].reshape((785, 1)), ei.reshape((1, 10)))
        w = w + Di
```

```python
print(w)

# 60000x10
y_1 = x@w
y_1 = make_answer(y_1)
np.set_printoptions(threshold=sys.maxsize)

# Выходные значения
print(y_1)
np.set_printoptions(threshold=10)
```

```python
images_test, labels_test = mnist_data.load_testing()

learning_data = mnist_data.process_images_to_numpy(images_test)

correct_answers = mnist_data.process_labels(labels_test)
```

```python
y = correct_answers
x = np.around(learning_data / 255)
x = np.hstack((x, np.ones((len(x), 1))))
# Тестовая выборка 10000 x 785
print(x)
```

```python
y_1 = x@w
y_1 = make_answer(y_1)
```

```python
def predict(vector):
    result = 0
    for i in range(10):
        if(vector[i] == 1): result = i
    return result
```

```python
right = 0
for i in range(len(x)):
    if(predict(y_1[i]) == y[i]):
        right += 1

accyracy = right / 100

# Правильных ответов в %
print(accyracy)
```
