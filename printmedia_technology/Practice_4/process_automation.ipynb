```python
import numpy as np
from scipy import ndimage
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# Оператор Робертса
roberts_cross_v = np.array([[0, 0, 0],
                            [0, 1, 0],
                            [0, 0, -1]])

roberts_cross_h = np.array([[0, 0, 0],
                            [0, 0, 1],
                            [0, -1, 0]])


def roberts_cross(img_data):
    vertical = ndimage.convolve(img_data, roberts_cross_v)
    horizontal = ndimage.convolve(img_data, roberts_cross_h)

    return np.sqrt(np.square(horizontal) + np.square(vertical))
```

```python
def light_gradation(light_arr, img_arr):
  light_arr = light_arr / img_arr.size

  tresh_hold = (np.max(light_arr) * 10) / 100
  temp_light_arr = light_arr > tresh_hold

  light_gradation_count = temp_light_arr.sum()

  if light_gradation_count > 7:
    print("Тоновое (Рекомендуется растрирование)")

  elif light_gradation_count <= 2 and light_gradation_count != 0:
    print("Штриховое")

  else:
    print("Многоградационное (Рекомендуется растрирование)")
```

```python
def check_color(img):
  img_RGB = img.convert('RGB')
  weight, height = img_RGB.size

  for i in range(weight):
    for j in range(height):
      r, g, b = img_RGB.getpixel((i, j))

      if r != g != b:
          return False

  return True
```

```python
def print_image(img, name):
    fig, axes = plt.subplots(1)

    axes.imshow(img, cmap="gray")
    axes.set_title(name)
    axes.axis("off")

    fig.set_figwidth(5)
    fig.set_figheight(5)

    plt.show()
```

```python
def print_chart(img_arr, text):
    x = [i for i in range(0, 256)]
    y = np.zeros(256, dtype=int)
    y1 = np.zeros(256, dtype=int)
    y2 = np.zeros(256, dtype=int)
    y3 = np.zeros(256, dtype=int)
    plt.figure(figsize=[5, 3])

    if len(img_arr.shape) > 2:
      for i in img_arr[:, :, 0].flat:
        y[i] += 1
      for i in img_arr[:, :, 1].flat:
        y1[i] += 1
      for i in img_arr[:, :, 2].flat:
        y2[i] += 1
      for i in img_arr.flat:
        y3[i] += 1
      plt.bar(x, y)
      plt.bar(x, y1)
      plt.bar(x, y2)

    else:
      for i in img_arr.flat:
        y3[i] += 1
      plt.bar(x, y3)

    plt.title(text)
    plt.ylabel("Кол-во пикселей")
    plt.xlabel("Уровень яркости")
    plt.xlim([-1, 256])
    plt.show()

    light_gradation(y3, img_arr)
```

```python
# Проход по всем изображениям в директории
for count in range(1, 31):
    image = Image.open(f'images/{count}.jpg')

    img_arr = np.asarray(image)

    # Определение средней яркости пикселей
    img_data = np.asarray(image.convert("L"), dtype="int32")
    avg_brightness = int(np.median(img_data))

    # Классификация изображения
    brightness = ""
    light_level = 0

    if(avg_brightness in range(130, 255)):
        brightness = "Светлое изображение"
        light_level = 0.78

    elif(avg_brightness in range(100,130)):
        brightness = "Среднее изображение"
        light_level = 0.46

    elif(avg_brightness in range(0, 100)):
        brightness = "Тёмное изображение"
        light_level = 0.11

    # Выделение контуров для определения занятой деталями площади
    border_mask = img_data > avg_brightness * light_level
    img_data[border_mask] = 255
    img_data[np.invert(border_mask)] = 0

    roberts_image_data = roberts_cross(img_data)

    # Сохранение полученных контуров для изображений в директорию
    Image.fromarray(np.uint8(Image.fromarray(roberts_image_data))).save(f'roberts/{count}.png')

    # Расчёт площади занятой деталями изображения по контуру
    tones_pixels_count = np.count_nonzero(roberts_image_data < 128)
    image_pixels_count = img_data.shape[0] * img_data.shape[1]
    tones_percent = tones_pixels_count / image_pixels_count

    # Вывод статистики по изображению
    print("Изображение: №", count)
    print_image(image, count)
    print_chart(img_arr, count)


    print(f"1) Яркость: {brightness}")
    print(f"2) Площадь, занятая равномерными тонами: {tones_percent}")

    if tones_percent < 0.94:
        print(f"3) Изображение с большим количеством сюжетных деталей")
    elif tones_percent < 0.96:
        print(f"3) Среднее содержание контуров")
    else:
        print(f"3) Преобладают равномерные тона")

    if check_color(image) == True:
        print("4) Черно-белое")
    else:
        print("4) Цветное (Рекомендуется цветоделение)")

    print("\n\n\n")
```
